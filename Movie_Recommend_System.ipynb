```python
"""

[1]  Demographic Filtering (인류통계학적 필터링)

* IDMb 메타스코어를 사용하여 높은 점수를 받은 영화를 목록으로 추천하는 시스템 
* A system that recommends movies with high scores as a list using IDMb metacorescores


"""
df1.columns = ['id','tittle','cast','crew']
df2= df2.merge(df1,on='id')
```

```python
df2.head(5)
```

```python
# 전체영화의 평균 획득 점수
C= df2['vote_average'].mean()
C
```

```python
# 90%이상의 득표수를 가진 영화 
m= df2['vote_count'].quantile(0.9)
m
```

```python
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
# IDMb 메타스코어 가중치 계산 함수 정의
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
# IDMb 메타스코어 가중치 내림차순 정렬 
q_movies = q_movies.sort_values('score', ascending=False)

q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
"""

[2]  Content Based Filtering (콘텐츠 기반 필터링)

* 사용자가 선택한 영화제목의 개요와 관련된 다른 영화 추천 시스템 
* A system for recommending movies associated with keywords in movies of your choice.


"""
# df2의 '개요' 컬럼 확인 
df2['overview'].head(5)
```

```python
# scikit-learn - TfIdfVectorizer imort
from sklearn.feature_extraction.text import TfidfVectorizer

# TfidVecrorizer 객체 정의. 모든 'the', 'a',등 어미 영단어 제거
tfidf = TfidfVectorizer(stop_words='english')

# NaN을 빈 문자열로 바꾸기
df2['overview'] = df2['overview'].fillna('')

# 데이터를 변환하여 적합하고 필요한 TF-IDF 행렬 구성
tfidf_matrix = tfidf.fit_transform(df2['overview'])

# 데이터셋 (영화의 수, 단어들의 수)
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

# 데이터셋의 코사인 유사도 계산 
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#인덱스 및 영화 제목 역시리즈 구성
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
```

```python
indices.head()
```

```python
# 가장 유사한 영화를 입력받아 추천하는 함수 정의 
def get_recommendations(title, cosine_sim=cosine_sim):
    
    # 입력받은 영화의 제목과 일치하는 인덱스 가져오기 
    idx = indices[title]
    
    # 인덱스로 모든 영화의 유사성 점수 계산
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 유사성 점수에 따른 영화 분류
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # 가장 유사한 점수 영화 10편 가져오기
    sim_scores = sim_scores[1:11]
    
    # 영화 인덱스 가져오기 
    movie_indices = [i[0] for i in sim_scores]
    
    # 가장 유사한 영화 10편 제목 반환 
    return df2['title'].iloc[movie_indices]
```

```python
"""

[2]  Content Based Filtering - 2

* 주연배우 3명, 감독, 장르등의 추가적인 키워드 사용 
* build a recommender based on the following metadata: 
  the 3 top actors, the director, related genres and the movie plot keywords.


"""
# 파이썬객체에 문자열화된 구문 분석 
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)
```

```python
# 제작진들중 특정 이름들을 가져오기. 존재하지 않는 경우 NaN 반환 
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# 리스트에서 상위 3명의 이름 반환. 혹은 3명 이상 
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        # 3개이상의 요소가 있는지 확인. 그렇다면 처음 3개만 반환. 아니라면 전부 반환 
        if len(names) > 3:
            names = names[:3]
        return names
    # 항목이 누락되거나 변형된 경우 빈 목록 변환
    return []
```

```python
# 적합한 형태의 새로운 특징들 정의 (제작진,캐스팅,장르...)
df2['director'] = df2['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
# 처음 3영화의 새로운 특징들 출력 
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
# 모든 문자열을 소문자로 변환 및 공백제거 함수 정의 
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        # 제작진 항목이 비어있다면, 공백 문자열 반환 
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
# 새로운 데이터셋을 함수에 적용 
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + \
           ' '.join(x['cast']) + ' ' + x['director'] + ' ' + \
           ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
```

```python
# TF-IDF 대신 CountVectorizer 사용
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
```

```python
# 코사인 유사도 계산 
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
# DataFrame의 기존 인덱스 초기화 및 시리즈 구성 
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title'])
```

```python

"""

[3] Collaborative Filtering (협업 필터링)

* 개선방안
* SVD (Single Value Decomposition) 이용 
* 차원 축소 알고리즘을 이용한 고차원 추천 시스템 구현 
* Implementation of High-Dimensional Recommendation System Using Dimension Reduction Algorithms


"""
#Surprise import
from surprise import Reader, Dataset, SVD, evaluate
reader = Reader()
ratings = pd.read_csv('data/ratings_small.csv')
ratings.head()
```

```python
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
# 교차검증을 위한 데이터분할 
data.split(n_folds=5)
```

```python
import random
import warnings
warnings.filterwarnings(action='ignore')
```

```python
svd = SVD()
# 평균 제곱근 편차, 평균 절대 오차 계산 
evaluate(svd, data, measures=['RMSE', 'MAE'])
```

```python
trainset = data.build_full_trainset()
# 데이터셋 학습시키기 
svd.fit(trainset)
```

```python
# SVD 예측 점수를 이용하여 비슷한 점수의 영화를 무작위로 추천하는 함수 정의 
def recommend(userid, movieid):
    
    # 예측 점수 소수점 첫째자리 변환 
    est = str(round((svd.predict(userid, movieid).est), 1))
    
    # 소수점 첫째자리가 0~2일경우 0으로 변환
    if 0 <= int(est[2]) <= 2:
        new_est = est[:2]+'0'
    # 3~7일 경우 5로 변환
    elif 3 <= int(est[2]) <= 7:
        new_est = est[:2]+'5'
    # 8~9일 경우 소수점 첫째자리에서 반올림 
    else:
        new_est = round((svd.predict(userid, movieid).est), 0)
        
    # 변환된 예측점수와 같은 점수를 가진 영화 데이터프레임 생성     
    raw_data = {'id': svd.predict(userid, movieid).uid,
                'title':df3[df3['rating'] == float(new_est)]['title']}
    recommend = pd.DataFrame(raw_data)
    
    # 중복을 제거하고 df에서 리스트로 영화제목들을 추출
    recommend = recommend.drop_duplicates()
    lst = list(recommend['title'])
    
    # 함수를 실행할때마다 리스트에서 랜덤으로 5개의 영화들을 출력
    return random.sample(lst, 5)
```

```python
# userID가 30인 사용자가 559영화를 선택했을때 예상되는 추천 영화 제목들 
recommend(30, 559)
```
